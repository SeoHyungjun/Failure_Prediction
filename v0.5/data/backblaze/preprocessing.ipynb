```python
import csv
import os
from random import *
import shutil

path_dir = '/Failure_Prediction/v0.4/data/backblaze/serial_sort/'
failure_dir = '/Failure_Prediction/v0.4/data/backblaze/serial_failure/'
normal_dir = '/Failure_Prediction/v0.4/data/backblaze/serial_normal/'

list_failure_dict = []
list_normal_file = []
list_failure_file = []

years = ['2018/', '2019/']
models = ['ST4000DM000/', 'ST12000NM007/']

for year in years:
    for model in models:
        file_list = os.listdir(path_dir + model + year)
        file_list.sort()

        failure_dict = {}
        failure_cnt = 0
        total_cnt = 0
        line_cnt = 0
        failure_file = []
        normal_file = []

        for file_name in file_list:
            with open(path_dir + model + year + file_name, 'r', encoding="utf-8") as f:        
                lines = list(csv.reader(f))
                line_cnt = 0
                failure_cnt = 0
                for line in lines:                    
                    if line[4] == '1':
                        failure_cnt += 1

                        if file_name not in failure_dict.keys():
                            failure_dict[file_name] = 1
                            failure_file.append(file_name)
                        else:
                            print('------- failure > 2   ' + file_name)
                            failure_dict[file_name] += 1

                if failure_cnt == 0:
                    normal_file.append(file_name)            

                total_cnt += 1
        
        for file in failure_file:
            shutil.copy(path_dir + model + year + file, failure_dir + model + year + file )

        for file in normal_file:
            shutil.copy(path_dir + model + year + file, normal_dir + model + year + file)
                
        list_failure_dict.append(failure_dict)
        list_normal_file.append(normal_file)
        list_failure_file.append(failure_file)
```
