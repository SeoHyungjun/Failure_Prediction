```python
import csv
import os
from random import *
```

```python
path_dir = '/gitlfs/Failure_Prediction/v0.5/data/backblaze/'
failure_dir = '/gitlfs/Failure_Prediction/v0.5/data/backblaze/serial_failure/'
normal_dir = '/gitlfs/Failure_Prediction/v0.5/data/backblaze/serial_normal/'
failure_save_dir = '/gitlfs/Failure_Prediction/v0.5/data/backblaze/backblaze_failure/'
normal_save_dir = '/gitlfs/Failure_Prediction/v0.5/data/backblaze/backblaze_normal/'

models = ['ST12000NM0007/'] #, 'ST4000DM000/']
years = ['2018/', '2019/']
```

```python
#4일(장애 발생일 + 3일) 까지 장애 데이터로 저장
days = 8
```

장애 데이터를 저장하는 부분

```python
for model in models:
    for year in years:
        file_list = os.listdir(failure_dir + model + year)
        file_list.sort()

        file_cnt = 0
        data_cnt = 0

        with open(failure_save_dir + '1819_12model_failure_' + str(days) + 'days.csv', 'a') as save_file:
            csvwriter = csv.writer(save_file)

            for file in file_list:
                file_cnt += 1
                with open(failure_dir + model + year + file, 'r') as f:
                    lines = list(csv.reader(f))

                    if lines[-1][4] == '1':
                        print(file + ' ' + str(len(lines)))

                        if len(lines) >= days:
                            for i in range(0, days):
                                lines[-1 - i][4] = 1
                                csvwriter.writerow(lines[-1 - i])
                                data_cnt += 1
                        else:
                            for i in range(0, len(lines)):
                                lines[-1 - i][4] = 1
                                csvwriter.writerow(lines[-1 - i])
                                data_cnt += 1
```

```python
path_dir = '/Failure_Prediction/v0.4/data/backblaze/'
failure_dir = '/Failure_Prediction/v0.4/data/backblaze/serial_failure/'
normal_dir = '/Failure_Prediction/v0.4/data/backblaze/serial_normal/'
failure_save_dir = '/Failure_Prediction/v0.4/data/backblaze/backblaze_failure/'
normal_save_dir = '/Failure_Prediction/v0.4/data/backblaze/backblaze_normal/'

models = ['ST12000NM0007', 'ST4000DM000']
years = ['2018', '2019']
t = '/'

for model in models:
    for year in years:
        print(year + ' ' + model)
        file_list = os.listdir(normal_dir + model + t + year + t)
        file_list.sort()

        file_cnt = 0
        data_cnt = 0

        with open(normal_save_dir + year + '_' + model + '_normal.csv', 'a') as save_file:
            csvwriter = csv.writer(save_file)

            for file in file_list:
                with open(normal_dir + model + t + year + t + file, 'r') as f:
                    lines = list(csv.reader(f))
                    for line in lines:
                        csvwriter.writerow(line)
```
